```python
import matplotlib.pyplot as plt
from bone_data.DataLoad import train_data, val_data, test_data
from PIL import Image
```

```python
#data load

train = train_data
val = val_data
test= test_data

train_img_path = 'bone_data/train/'
validation_img_path = 'bone_data/validation/'
test_img_path = 'bone_data/test/'
```

```python
train_list_x, train_list_y, val_list_x, val_list_y, test_list_x, test_list_y =[], [], [], [], [], []

for i in range(0,12611,1):
    name = train_img_path + str(train.iloc[i, 0]) + '.png'
    img = Image.open(name)
    train_list_x.append(img.size[0])
    train_list_y.append(img.size[1])


for i in range(0,1425,1):
    name = validation_img_path + str(val.iloc[i, 0]) + '.png'
    img = Image.open(name)
    val_list_x.append(img.size[0])
    val_list_y.append(img.size[1])

for i in range(0,200,1):
    name = test_img_path + str(test.iloc[i, 0]) + '.png'
    img = Image.open(name)
    test_list_x.append(img.size[0])
    test_list_y.append(img.size[1])
```

```python
plt.title('Trainset-Size-Distribution')
plt.scatter(train_list_x, train_list_y)
plt.axis('equal')
plt.xlabel('width')
plt.ylabel('height')
# plt.savefig('trainset-size-distribution.png', facecolor='#eeeeee')
plt.show()

plt.title('Validationset-Size-Distribution')
plt.scatter(val_list_x, val_list_y)
plt.axis('equal')
plt.xlabel('width')
plt.ylabel('height')
# plt.savefig('validationset-size-distribution.png', facecolor='#eeeeee')
plt.show()

plt.title('Testset-Size-Distribution')
plt.scatter(test_list_x, test_list_y)
plt.axis('equal')
plt.xlabel('width')
plt.ylabel('height')
# plt.savefig('testset-size-distribution.png', facecolor='#eeeeee')
plt.show()
```
